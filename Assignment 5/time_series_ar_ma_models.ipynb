# Time Series Analysis: AR and MA Models

This notebook provides an overview and visualizations of Autoregressive (AR) and Moving Average (MA) models, which are fundamental to understanding and forecasting time series data.

## Autoregressive (AR) Model

The Autoregressive (AR) model is a type of stochastic time series model that uses dependencies between an observed value and a number of lagged observations of itself. The AR model is defined by the equation:

$$X_t = c + \phi_1X_{t-1} + \phi_2X_{t-2} + ... + \phi_pX_{t-p} + \epsilon_t$$

where:
- $X_t$ is the value at time $t$.
- $c$ is a constant.
- $\phi_1, \phi_2, ..., \phi_p$ are the parameters of the model.
- $\epsilon_t$ is white noise.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.tsa.arima_process import ArmaProcess

# Simulate an AR(1) process
np.random.seed(123)
ar = np.array([1, -0.5]) # AR parameter
ma = np.array([1])       # MA parameter is set to 1 for AR model
AR_object = ArmaProcess(ar, ma)
simulated_data = AR_object.generate_sample(nsample=100)

# Plot the simulated AR(1) process
plt.figure(figsize=(10, 6))
plt.plot(simulated_data)
plt.title('Simulated AR(1) Process')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```

## Moving Average (MA) Model

The Moving Average (MA) model is another type of time series model that uses the dependency between an observation and a residual error from a moving average model applied to lagged observations. The MA model is defined by the equation:

$$X_t = \mu + \epsilon_t + \theta_1\epsilon_{t-1} + \theta_2\epsilon_{t-2} + ... + \theta_q\epsilon_{t-q}$$

where:
- $X_t$ is the value at time $t$.
- $\mu$ is the mean of the series.
- $\epsilon_t$ is white noise.
- $\theta_1, \theta_2, ..., \theta_q$ are the parameters of the model.

```python
# Simulate an MA(1) process
np.random.seed(123)
ar = np.array([1])       # AR parameter is set to 1 for MA model
ma = np.array([1, 0.5])  # MA parameter
MA_object = ArmaProcess(ar, ma)
simulated_data = MA_object.generate_sample(nsample=100)

# Plot the simulated MA(1) process
plt.figure(figsize=(10, 6))
plt.plot(simulated_data)
plt.title('Simulated MA(1) Process')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```
